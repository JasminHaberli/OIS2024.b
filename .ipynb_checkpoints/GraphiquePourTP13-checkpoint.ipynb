```r
# Charger le package ggplot2
library(ggplot2)

# Utiliser le jeu de données iris
df <- iris

# Créer la visualisation
p <- ggplot(df, aes(x = Sepal.Length, fill = Species)) +
  geom_density(alpha = 0.5) +  # Tracer des densités pour chaque espèce
  labs(title = "Comparaison des longueurs des sépales entre trois espèces d'iris",
       x = "Longueur du sépale (cm)",
       y = "Densité") +
  theme_minimal() +
  scale_fill_manual(values = c("#619CFF", "#FF8474", "#50D050"))  # Choisir des couleurs personnalisées

# Afficher la visualisation
print(p)
```
