```r
# Charger le package ggplot2
library(ggplot2)

# Utiliser le jeu de donnes iris
df <- iris

# Crer la visualisation
p <- ggplot(df, aes(x = Sepal.Length, fill = Species)) +
  geom_density(alpha = 0.5) +  # Tracer des densits pour chaque espce
  labs(title = "Comparaison des longueurs des spales entre trois espces d'iris",
       x = "Longueur du spale (cm)",
       y = "Densit") +
  theme_minimal() +
  scale_fill_manual(values = c("#619CFF", "#FF8474", "#50D050"))  # Choisir des couleurs personnalises

# Afficher la visualisation
print(p)
```
